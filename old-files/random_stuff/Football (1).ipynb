```python
train['market_value'] = train['market_value'].astype('int64')
train['fpl_value'] = train['fpl_value'].astype('int64')
display(train.dtypes)
```

```python
X = train.drop(['market_value','position_cat','age'], axis = 1)
y = train['market_value']
```

```python
from sklearn.model_selection import train_test_split as split
```

```python
X_train, X_test, y_train, y_test = split(X, y, test_size=0.25, random_state=0)
print(X_train)
print(y_train)
print(X_test)
```

```python
from sklearn import linear_model
from sklearn.metrics import accuracy_score
reg = linear_model.LinearRegression()
reg.fit(X_train,y_train)
coefficients = reg.coef_
y_pred = reg.predict(X_train)
print('Coefficients: ', coefficients)
print('Linear Regression R^2 score on train data = ' , round(reg.score(X_train,y_pred) * 100, 2))
print('Linear Regression R^2 score on test data = ' , round(reg.score(X_test,y_test) * 100, 2))
print(y_pred)
```

```python
from sklearn.linear_model import Lasso
reg = Lasso()
reg.fit(X_train,y_train)
coefficients = reg.coef_

print('Coefficients: ', coefficients)
print('Lasso Regression R^2 score on train data = ' , round(reg.score(X_train,y_pred) * 100, 2))
print('Lasso Regression R^2 score on test data = ' , round(reg.score(X_test,y_test) * 100, 2))
y_pred = reg.predict(X_train)
y_pred
```

```python
from sklearn.linear_model import Ridge
reg = Ridge()
reg.fit(X_train, y_train)
coefficients = reg.coef_
y_pred = reg.predict(X_train)


print ('Coefficients:', coefficients)
print ('Ridge Regression: R^2 score on training set', round(reg.score(X_train,y_pred)*100,2))
print ('Ridge Regression: R^2 score on test set', round(reg.score(X_test,y_test)*100,2))
```

```python
from sklearn.neighbors import KNeighborsRegressor as KNeighbors
reg = KNeighbors()
reg.fit(X_train, y_train)
y_pred = reg.predict(X_train)
print ('K-Neighbor Regression: R^2 score on training set', round(reg.score(X_train,y_pred)*100,2))
print ('K-Neighbor Regression: R^2 score on test set', round(reg.score(X_test,y_test)*100,2))
```

```python
from sklearn.svm import SVR
reg = SVR()
reg.fit(X_train, y_train)
y_pred = reg.predict(X_train)

print ('K-Neighbor Regression: R^2 score on training set', round(reg.score(X_train,y_pred)*100,2))
print ('K-Neighbor Regression: R^2 score on test set', round(reg.score(X_test,y_test)*100,2))
```

```python
from sklearn.tree import DecisionTreeRegressor as dtr
reg = dtr()
reg.fit(X_train,y_train)

print ('K-Neighbor Regression: R^2 score on training set', round(reg.score(X_train,y_train)*100,2))
print ('K-Neighbor Regression: R^2 score on test set', round(reg.score(X_test,y_test)*100,2))
```

```python
from sklearn.ensemble import RandomForestRegressor as rfg
reg = rfg()
reg.fit(X_train, y_train)

print ('K-Neighbor Regression: R^2 score on training set', round(reg.score(X_train,y_train)*100,2))
print ('K-Neighbor Regression: R^2 score on test set', round(reg.score(X_test,y_test)*100,2))
```

```python
from sklearn.ensemble import GradientBoostingRegressor as gbr
reg = gbr()
reg.fit(X_train, y_train)
y_pred = reg.predict(X_train)
print ('K-Neighbor Regression: R^2 score on training set', round(reg.score(X_train,y_pred)*100,2))
print ('K-Neighbor Regression: R^2 score on test set', round(reg.score(X_test,y_test)*100,2))
y_pred
```
